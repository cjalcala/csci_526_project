```python
import pandas 
import json
df = pandas.read_excel('Questions11262022.xlsx',usecols=['Question','Answer','Other Options: B','Other Options: C','Other Options: D','Level Index','Hint'])

#url = "https://docs.google.com/spreadsheets/d/13Z-lnwtksOeN7GJNu3d9pgsnTTKFX6cIGgoiOLTdFeA/edit#gid=0"
#df = pandas.read_excel(url,usecols=['Question','Answer','Other Options: B','Other Options: C','Other Options: D','Level Index'])
# print whole sheet data
print(df)
```

```python
df = df.rename(columns={"Question": "QuestionText", "Answer": "Option1","Other Options: B": "Option2","Other Options: C": "Option3","Other Options: D": "Option4"})
```

```python
level1 = df[df['Level Index'] == 0].drop('Level Index', axis=1)
level2 = df[df['Level Index'] == 1].drop('Level Index', axis=1)
level3 = df[df['Level Index'] == 2].drop('Level Index', axis=1)
```

```python
questionDict = {"Level1":level1.to_dict(orient='records'),'Level2':level2.to_dict(orient='records'),'Level3':level3.to_dict(orient='records')}

print(questionDict)
```

```python
with open("QuestionsWithHints.json", "w",encoding="utf8") as outfile:
    #json.dump(df.to_dict(orient='records'), outfile,ensure_ascii=False)
    json.dump(questionDict, outfile,ensure_ascii=False)
outfile.close()
```
